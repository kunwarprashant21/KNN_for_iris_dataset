```python
import csv
import random
# spiliting the dataset into training set and test set
def dataset(filename, split, trainingset=[], testset=[]):
    with open(filename,'r') as csvfile:
        lines=csv.reader(csvfile)
        dataset=list(lines)
        for x in range(len(dataset)-1):
                for y in range(4):
                    dataset[x][y]=float(dataset[x][y])
                if random.random()<split:
                    trainingset.append(dataset[x])
                else:
                    testset.append(dataset[x])
```

```python
# printing the number of datapoints in each dataset
trainingset=[]
testset=[]
dataset('iris.csv', 0.70, trainingset, testset)
print('train: '+ repr(len(trainingset)))
print('test: '+ repr(len(testset)))
```

```python
# spliting the data and labels 
trainingset_data=[]
for x in range(len(trainingset)):
        trainingset_data.append(trainingset[x][0:4])

trainingset_label=[]
for x in range(len(trainingset)):
    trainingset_label.append(trainingset[x][4])
  
```

```python
# spliting the data and labels 

testset_data=[]
for x in range(len(testset)):
        testset_data.append(testset[x][0:4])

testset_label=[]
for x in range(len(testset)):
    testset_label.append(testset[x][4])
 
```

```python

import math
import numpy as np
# computing euclidean distances between two given points

def euclideandistance(instance1, instance2):
        neighbors= np.sqrt(np.sum(pow(np.array(instance1)-np.array(instance2)[:],2),1))
        indices=sorted(range(len(neighbors)), key=lambda k: neighbors[k])
        return indices
```

```python
import operator

# finding k nearest neighbours for each point of testdata 
def find_neighbors(trainingset_data, trainingset_label,testset_data,k):
    neighbors=[]
    
    for x in range(len(testset_data)):
        setosa=0
        versi=0
        verg=0
        neighbor_index=euclideandistance(trainingset_data, testset_data[x])
        for x in range(k):
            if trainingset_label[neighbor_index[x]]=='Iris-setosa':
                setosa=setosa+1
            elif trainingset_label[neighbor_index[x]]=='Iris-versicolor':   
                versi=versi+1
            else:
                verg=verg+1
        if setosa>versi:
            if setosa>verg:
                n='Iris-setosa'
            else:
                n='Iris-virginica'
        elif versi>verg:
            n='Iris-versicolor'
        else:
            n='Iris-virginica'
            
            
        neighbors.append(n)
        
    return neighbors
```

```python
def main():
    trainingset=[]
    testset=[]
    split=0.70
    dataset('iris.csv', split, trainingset, testset)
    print('train set:' + repr(len(trainingset)))
    print('test set:' + repr(len(testset)))
    predictions=[]
    k=5
    
    predictions=find_neighbors(trainingset_data,trainingset_label,testset_data,k)
    
    for x in range(len(predictions)):
        print('>predicted='+repr(predictions[x])+'  ,actual='+repr(testset_label[x]))
    correct=0
    for x in range(len(predictions)):
        if predictions[x]==testset_label[x]:
            correct+=1
    accuracy=correct/len(testset_label)*100
    print(accuracy)
main()
    
    
```
